```python
credits_column_renamed = credits.rename(index=str, columns= {"movie_id": "id"})
```

```python
movies_df_merge = movies_df.merge(credits_column_renamed, on = 'id')
```

```python
movies_cleaned_df = movies_df_merge.drop(columns = ['homepage', 'title_x', 'title_y', 'status', 'production_companies', 'crew', 'production_countries'])
movies_cleaned_df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df =3, max_features = None, strip_accents = 'unicode', analyzer = 'word', token_pattern = r'\w{1,}', ngram_range = (1,3),
                      stop_words = 'english')
```

```python
# filling Nans with empty string
movies_cleaned_df['overview'] = movies_cleaned_df['overview'].fillna('')
```

```python
# fitting the tfidf on the 'overview' text
tfv_matrix = tfv.fit_transform(movies_cleaned_df['overview'])
tfv_matrix
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# compute sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
sig[0]
```

```python
# Reverse mapping of indices and movie titles
indices = pd.Series(movies_cleaned_df.index, index=movies_cleaned_df['original_title']).drop_duplicates()
```

```python
def give_rec(title, sig=sig):
  # get the index corresponding to the original title
  idx = indices[title]

  # get the pairwise similarity index
  sig_scores = list(enumerate(sig[idx]))

  # sort the movies
  sig_scores = sorted(sig_scores, key=lambda x:x[1], reverse=True)

  # scores of the 10 most similar movies
  sig_scores = sig_scores[1:11]

  # movie indices
  movie_indices = [i[0] for i in sig_scores]

  # Top 10 most similar movies
  return movies_cleaned_df['original_title'].iloc[movie_indices]
```

```python
# Testing content based recommendation ssytem with the seminal film spy kids
give_rec('Spy Kids')
```
