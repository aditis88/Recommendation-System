**To ensure statistical significance, users with less than 200 ratings, and books with less than 100 ratings are excluded.**


---



```python
counts1 = ratings['user_id'].value_counts()
ratings = ratings[ratings['user_id'].isin(counts1[counts1 >= 200].index)]
counts = ratings['bookRating'].value_counts()
ratings = ratings[ratings['bookRating'].isin(counts[counts >= 100].index)]
```

```python
print("ratings columns:", list(ratings.columns))
print("books columns:", list(books.columns))
print("users columns:", list(users.columns))
```

```python
books.columns = books.columns.str.strip().str.lower()
ratings.columns = ratings.columns.str.strip().str.lower()

books.rename(columns={'isbn': 'ISBN'}, inplace=True)
ratings.rename(columns={'isbn': 'ISBN'}, inplace=True)

```

```python
print("After renaming:")
print("ratings columns:", list(ratings.columns))
print("books columns:", list(books.columns))
```

```python
combine_book_rating = pd.merge(ratings, books, on='ISBN')
columns = ['yearofpublication', 'publisher', 'bookauthor', 'imageurls', 'imageurlm', 'imageurll']

combine_book_rating = combine_book_rating.drop(columns, axis=1)
combine_book_rating.head()
```

**Collaborative Filtering Using k-Nearest Neighbors (kNN)**


---



```python
combine_book_rating = combine_book_rating.dropna(axis = 0, subset = ['booktitle'])

book_ratingCount = (combine_book_rating.
     groupby(by = ['booktitle'])['bookrating'].
     count().
     reset_index().
     rename(columns = {'bookrating': 'totalRatingCount'})
     [['booktitle', 'totalRatingCount']]
    )
book_ratingCount.head()
```

```python
rating_with_totalRatingCount = combine_book_rating.merge(book_ratingCount, left_on = 'booktitle', right_on = 'booktitle', how = 'left')
rating_with_totalRatingCount.head()
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
print(book_ratingCount['totalRatingCount'].describe())
```

```python
print(book_ratingCount['totalRatingCount'].quantile(np.arange(.9, 1, .01)))
```

```python
popularity_threshold = 50
rating_popular_book = rating_with_totalRatingCount.query('totalRatingCount >= @popularity_threshold')
rating_popular_book.head()
```

```python
rating_popular_book.shape
```

```python
#ratings.rename(columns={'user_id': 'userID'}, inplace=True)
```

```python
# Convert userID in both DataFrames to string
rating_popular_book['user_id'] = rating_popular_book['user_id'].astype(str)
users['userID'] = users['userID'].astype(str)
users.rename(columns={'userID': 'user_id'}, inplace=True)
# Now merge safely
combined = rating_popular_book.merge(users, on='user_id', how='left')

# Filter for USA/Canada
us_canada_user_rating = combined[combined['Location'].str.contains("usa|canada", case=False, na=False)]

# Drop age if needed
us_canada_user_rating = us_canada_user_rating.drop('Age', axis=1)

# Preview
us_canada_user_rating.head()
```

```python
from scipy.sparse import csr_matrix
us_canada_user_rating = us_canada_user_rating.drop_duplicates(['user_id','booktitle'])
us_canada_user_rating_pivot = us_canada_user_rating.pivot(index='booktitle', columns='user_id', values='bookrating').fillna(0)
us_canada_user_rating_matrix = csr_matrix(us_canada_user_rating_pivot.values)
```

```python
us_canada_user_rating_pivot.head()
```

```python
from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(us_canada_user_rating_matrix)
```

```python
query_index = 5
us_canada_user_rating_pivot.iloc[query_index, :].values.reshape(1, -1)
```

```python
query_index = np.random.choice(us_canada_user_rating_pivot.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(us_canada_user_rating_pivot.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
us_canada_user_rating_pivot.index[query_index]
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(us_canada_user_rating_pivot.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, us_canada_user_rating_pivot.index[indices.flatten()[i]], distances.flatten()[i]))
```
